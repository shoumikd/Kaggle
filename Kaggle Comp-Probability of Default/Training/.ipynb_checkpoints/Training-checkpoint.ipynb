```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
credit_risk_orig=pd.read_csv('./data/credit_risk_dataset.csv')
credit_risk_orig.head()
```

```python
credit_risk_orig.describe()
```

```python
cr_copy=credit_risk_orig.copy()
```

```python
#cr_copy.pivot_table( index=['person_age'],columns=['loan_status'],values=['person_income'],aggfunc="count").sort_index()
age_filter=cr_copy['person_age']<=68
cr_age_filtered=cr_copy[age_filter]
cr_age_filtered.shape
```

```python
#cr_copy.pivot_table( index=['person_emp_length'],columns=['loan_status'],values=['person_income'],aggfunc="count").sort_index()
emp_length_filter=cr_age_filtered['person_emp_length']<=47
cr_emp_length_filtered=cr_age_filtered[emp_length_filter]
cr_emp_length_filtered.shape
cr_emp_length_filtered.head()
```

```python
cr_emp_length_filtered.describe()
```

```python
cr_emp_length_filtered.reset_index(drop=True,inplace=True)
cr_emp_length_filtered.head()
```

```python
cr_emp_length_filtered.isnull().sum()
```

```python
cr_int_rate_filled=cr_emp_length_filtered.copy()
cr_int_rate_filled=cr_int_rate_filled.fillna({'loan_int_rate':cr_int_rate_filled['loan_int_rate'].mean()})
cr_int_rate_filled.describe()
cr_int_rate_filled.head()
```

```python
cr_int_rate_filled.groupby('loan_status').count()['person_age']
```

```python
cr_loan_grade_rmvd=cr_int_rate_filled.drop('loan_grade',axis=1)
cr_loan_grade_rmvd.shape
```

```python
cr_loan_grade_rmvd.head()
```

### Categorical Feature Treatment

```python
cr_loan_cat_treated=cr_loan_grade_rmvd.copy()
cr_loan_cat_treated['person_home_ownership'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

cr_loan_cat_treated.dtypes
categorical_columns=cr_loan_cat_treated.select_dtypes(include=['object']).columns.to_list
print(categorical_columns)

for column in categorical_columns():
    print (cr_loan_cat_treated[column].value_counts())

one_hot_encoding_columns=['person_home_ownership', 'loan_intent']
binary_encoding_columns=['cb_person_default_on_file']

#Initialize One Hot Encoder
ohe=OneHotEncoder(sparse_output=False)

#Apply one-hot encoder to categorical columns
one_hot_encoded=ohe.fit_transform(cr_loan_cat_treated[one_hot_encoding_columns])

#create dataframe for categorical variables
cr_ohe=pd.DataFrame(one_hot_encoded, columns=ohe.get_feature_names_out(one_hot_encoding_columns))

#merge parent dataframe with categorical features dataframe 
cr_loan_cat_treated = pd.concat([cr_loan_cat_treated, cr_ohe], axis=1)
cr_loan_cat_treated.shape
```

```python
#populate binary columns with 0 and 1
cr_loan_bin_treated=cr_loan_cat_treated.copy()
cr_loan_bin_treated[binary_encoding_columns]=np.where((cr_loan_bin_treated[binary_encoding_columns])=='Y',1,0)
cr_loan_bin_treated.head()
```

```python
cr_loan_bin_treated.dtypes
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

scale_columns=['person_age','person_income','person_emp_length','loan_amnt','loan_int_rate','loan_percent_income','cb_person_cred_hist_length']

cr_loan_scaled=cr_loan_bin_treated.copy()
cr_loan_scaled[scale_columns] = StandardScaler().fit_transform(cr_loan_scaled[scale_columns])
cr_loan_scaled.head()
```

```python
dropColumns=['person_home_ownership','loan_intent','person_home_ownership_OTHER','loan_intent_DEBTCONSOLIDATION']

cr_loan_processed=cr_loan_scaled.copy()
cr_loan_processed= cr_loan_processed.drop(dropColumns, axis=1)
cr_loan_processed.shape
# cr_loan_processed.head()
```

```python
cr_loan_processed['loan_status'].value_counts()
```

### SMOT-Synthetic Minority Over-Sampling Technique

```python
from imblearn.over_sampling import SMOTE

smote=SMOTE()

independent=cr_loan_processed.drop('loan_status', axis=1)
dependent=cr_loan_processed['loan_status']

x_smote, y_smote=smote.fit_resample(independent,dependent)
x_smote.head()
```

```python
y_smote.value_counts()
```

### Train Test Data

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

x_train, x_test, y_train, y_test = train_test_split(x_smote, y_smote, test_size=0.2, random_state=42)

x_train.head()
```

### Model Creation

```python
from sklearn.linear_model import LogisticRegression
import pickle

logit= LogisticRegression()

logit.fit(x_train, y_train)

with open('logistic_pd_model.pkl','wb') as file:
    logistic_pd_model=pickle.dump(logit,file)
    
logit.score(x_train, y_train)
```

```python
logit_pred=logit.predict(x_test)
logit_pred
```

```python
print(classification_report(y_test,logit_pred))
print(logit.coef_)
```

```python
feature_logit_imp=pd.DataFrame({'features':x_smote.columns,'logit_impr':logit.coef_[0]})
feature_logit_imp.sort_values('logit_impr',ascending=False)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf= RandomForestClassifier()

rf.fit(x_train, y_train)

with open('rf_pd_model.pkl','wb') as file:
    rf_pd_model=pickle.dump(rf,file)

rf_pred=rf.predict(x_test)

rf.score(x_train, y_train)
rf_pred
```

```python
print(classification_report(y_test,rf_pred))
print(rf.feature_importances_)
```

```python
feature_rf_imp=pd.DataFrame({'features':x_smote.columns,'rf_impr':rf.feature_importances_})
feature_rf_imp.sort_values('rf_impr',ascending=False)
```

```python
from xgboost import XGBClassifier

xgb=XGBClassifier(tree_method='exact')
xgb.fit(x_train,y_train.values.ravel())

with open('xgb_pd_model','wb') as file:
    xgb_pd_model=pickle.dump(xgb,file)

xgb.score(x_train,y_train.values.ravel())
```

```python
xgb_pred=xgb.predict(x_test)
```

```python
print(classification_report(y_test,xgb_pred))
print(xgb.feature_importances_)
```

```python
feature_xgb_imp=pd.DataFrame({'features':x_smote.columns,'xgb_impr':xgb.feature_importances_})
feature_xgb_imp.sort_values('xgb_impr',ascending=False)
```

```python
feature_imp=feature_logit_imp.merge(right=feature_rf_imp,on='features').merge(right=feature_xgb_imp,on='features')
feature_imp
```
