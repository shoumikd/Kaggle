```python
import pandas as pd
import numpy as np
#from llama_index import SimpleDocument, GPTListIndex
import os
```

### Load the dataset

```python
file_path = "..\\Training\\data\\credit_risk_dataset.csv"
print("Absolute Path:",os.path.abspath(file_path))

try:
    with open(file_path) as f:
        df = pd.read_csv(file_path)
except FileNotFoundError:
    raise FileNotFoundError('The file does not exist.')
```

### Set Target Column

```python
target_column='loan_status'
```

### Retrieve Numeric and Categorical Columns

```python
numerical_columns=df.select_dtypes(include=np.number).columns.tolist()
print(numerical_columns)
categorical_columns=df.select_dtypes(include=['category', 'object']).columns.tolist()
print(categorical_columns)
```

```python
df.describe(include='all')
```

```python
# 1. Drop Duplicates




# 2. Handle missing values
# Impute missing values in 'loan_int_rate' with the median
df['loan_int_rate'].fillna(df['loan_int_rate'].median(), inplace=True)

# 3. Convert columns to categorical types
for col in categorical_columns:
    df[col] = df[col].astype('category')

# 4. Outlier Detection and Removal
# Remove outliers in 'person_age' using the IQR method
Q1 = df['person_age'].quantile(0.25)
Q3 = df['person_age'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['person_age'] < (Q1 - 1.5 * IQR)) | (df['person_age'] > (Q3 + 1.5 * IQR)))]

# Remove outliers in 'person_income' using the IQR method
Q1 = df['person_income'].quantile(0.25)
Q3 = df['person_income'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['person_income'] < (Q1 - 1.5 * IQR)) | (df['person_income'] > (Q3 + 1.5 * IQR)))]

# Similarly for 'cb_person_cred_hist_length'
Q1 = df['cb_person_cred_hist_length'].quantile(0.25)
Q3 = df['cb_person_cred_hist_length'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['cb_person_cred_hist_length'] < (Q1 - 1.5 * IQR)) | (df['cb_person_cred_hist_length'] > (Q3 + 1.5 * IQR)))]

# 5. Summarize the cleaned dataset using LlamaIndex
#document = SimpleDocument.from_dataframe(df)
#index = GPTListIndex(documents=[document])

# Get a summary using Llama
#summary = index.query("Summarize the cleaned dataset statistics and data types.")

# Display the summary
#print(summary)

# Save the cleaned data
df.to_csv('cleaned_predicted.csv', index=False)
```
